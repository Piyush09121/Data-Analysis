```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

list = ["CRIM","ZN", "INDUS", "CHAS","NOX","RM","AGE","DIS","RAD", "TAX", "PTRATIO","B","LSTAT","MEDV"]
df = pd.read_csv(r"C:/functionup assignments/python project/housing.csv",header=None,delimiter=r"\s+",names=list)
df_copy= df.copy()
df_copy
```

```python
df_copy
```

```python
df_copy.corr()
```

```python

sns.histplot(data=df_copy, x='CRIM')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_copy, x='RM', y='MEDV')
plt.xlabel('Average Number of Rooms (RM)')
plt.ylabel('Median Home Value (MEDV)')
plt.title('Relationship between Rooms and Home Value')
plt.show()




```

```python

plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='CHAS', y='MEDV')
plt.xlabel('Bound Charles River (CHAS)')
plt.ylabel('Median Home Value (MEDV)')
plt.title('Median Home Value by Charles River Bounding')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.boxplot(data=df_copy, x='CRIM')
plt.xlabel('Per Capita Crime Rate (CRIM)')
plt.title('Distribution of Crime Rate')
plt.show()
```

```python

correlation_matrix = df_copy.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.countplot(data=df_copy, x='CHAS')
plt.xlabel('Bound Charles River (CHAS)')
plt.ylabel('Frequency')
plt.title('Frequency of Charles River Bounding')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='CHAS', y='MEDV')
plt.xlabel('Bound Charles River (CHAS)')
plt.ylabel('Median Home Value (MEDV)')
plt.title('Distribution of Home Values by Charles River Bounding')
plt.show()




```
